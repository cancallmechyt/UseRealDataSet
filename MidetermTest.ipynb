```python
import pandas as pd
df = pd.read_csv('./files/BurgerKing_Menu.csv')
df = df[['Menu', 'Category', 'Calories', 'FatCalories', 'Fat', 'SaturatedFat', 
         'Cholesterol', 'Sodium', 'TotalCarb', 'Sugars', 'Protein', 'DietaryFiber']] ##เพื่อเลือกและเก็บเฉพาะคอลัมน์ที่ระบุ
df
```

```python
##แปลงข้อมูลจาก DataFrame (df) เป็นรายการของ tuples
bkdb = [tuple(row) for row in df.itertuples(index=False)]
bkdb

cursor = connection.cursor()
cursor.execute("DESCRIBE BurgerKingMenuDB")

results = cursor.fetchall()
for row in results:
    print(row)
```

```python
bkdb = connection.cursor()

insert_sql = '''
                INSERT INTO BurgerKingMenuDB(Menu, Category, Calories, FatCalories, Fat, SaturatedFat, Cholesterol, Sodium, TotalCarb, Sugars, Protein, DietaryFiber)
                VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
            '''

values_to_insert = [tuple(row) for row in df.values]

bkdb.executemany(insert_sql, values_to_insert)
connection.commit()

print(bkdb.rowcount)
```

### กราฟที่ 0 แสดงเมนูที่มีจำนวนแคลอรี่สูงที่สุดจากข้อมูลทั้งหมด
#### ► เหตุผลที่เลือก Scatterplot เพราะ เหมาะกับการแสดงข้อมูลที่มีจำนวนมาก 

```python
import matplotlib.pyplot as plt

##เรียงลำดับข้อมูลใน DataFrame ที่มีค่าสูงสุดไปหาต่ำสุด และเก็บผลลัพธ์ลงในตัวแปร df_sorted
df_sorted = df.sort_values(by='Calories', ascending=False)

plt.figure(figsize=(20, 10))
##สร้าง scatter plot โดยใช้ข้อมูลจากคอลัมน์ 'Menu'                                   (ขนาดของจุด)
plt.scatter(df_sorted['Menu'], df_sorted['Calories'], color='skyblue', marker='o', s=100, alpha=0.7, label='Calories')
##กำหนดชื่อ title
plt.title('Calories in Menu Items (Highest to Lowest)')

##กำหนดป้ายกำกับในแกน x และแกน y
plt.xlabel('Menu') 
plt.ylabel('Calories')
plt.xticks(rotation=90)

plt.grid(True, linestyle='--', alpha=0.5) ##กำหนดรูปแบบของเส้นกริดเป็นเส้นประ 
plt.legend() ##ปรับระยะห่างของข้อมูลในกราฟ
plt.tight_layout()
plt.show()
```

## กราฟที่ 1 แสดงเมนูที่มี Sodium, Calories, Cholesterol, Protein มากที่สุด 30 รายการ

#### ► เหตุผลที่เลือก Barplot(แบบซ้อนกัน) เพราะ สามารถแสดงข้อมูลของ Sodium, Calories, Cholesterol, Protein ในเมนได้อย่างชัดเจน

```python
import matplotlib.pyplot as plt

plt.style.use('bmh')
##เลือกค่า 30 รายการที่มีค่าสูงสุดในคอลัมน์ 'Calories' และเก็บไว้ในตัวแปร 'top_10_items' (nlargest ช่วยในการเรียงลำดับรายการตามค่าสูงสุด)
top_10_items = df.nlargest(30, 'Calories')

plt.figure(figsize=(12, 6))
##สร้าง plotbar 
plt.bar(top_10_items['Menu'], top_10_items['Sodium'], label='Sodium', alpha=0.7)
plt.bar(top_10_items['Menu'], top_10_items['Calories'], label='Calories', alpha=0.7)
plt.bar(top_10_items['Menu'], top_10_items['Cholesterol'], label='Cholesterol', alpha=0.7)
plt.bar(top_10_items['Menu'], top_10_items['Protein'], label='Protein', alpha=0.7)

plt.title('Top 30 Menu with Highest (Calories, Sodium, Cholesterol, Protein)\n Sort by highest calorie')
plt.xlabel('Menu')
plt.ylabel('Values')

plt.legend()
plt.xticks(rotation=90) 

plt.show()
```

## กราฟที่ 2 แสดงหมวดหมู่เบอร์เกอร์ที่มี Calories และ Fat Calories มากที่สุด 12 รายการ
### ► เหตุผลที่เลือก Bar Plot (Multi) เพราะ สามารถแสดงข้อมูล Calories และ Fat Calories ได้อย่างชัดเจน

```python
import matplotlib.pyplot as plt

##กรองข้อมูลใน DataFrame "df" เพื่อเลือกเฉพาะรายการอาหารที่อยู่ในหมวดหมู่ "Burgers
df_filtered = df[df['Category'] == 'Burgers'].nlargest(12, 'Calories')

##สร้างกราฟในขนาด 12x12 และแบ่งให้มีแกนสองแกน โดย ax1 เป็นแกนหลักและ ax2 เป็นแกนรอง
fig, ax1 = plt.subplots(figsize=(12, 12))

##สร้างกราฟแท่งบน ax1 โดยใช้ข้อมูลจากคอลัมน์ 'Menu' เป็นแกน x และคอลัมน์ 'Calories' เป็นแกน y
ax1.bar(df_filtered['Menu'], df_filtered['Calories'], color='b', alpha=0.3, label='Calories')
ax1.set_ylabel('Calories', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_title('Top 12 High-Calorie and Fat Calories in Menu Category Burgers')

##สร้างแกนรอง ax2 ที่ใช้เลขความสูงเดียวกับ ax1.
ax2 = ax1.twinx()
bar_width = 0.3
index = range(len(df_filtered))

##สร้างกราฟแท่งบน ax2 โดยใช้ข้อมูลจากคอลัมน์ 'FatCalories' 
ax2.bar([i + bar_width for i in index], df_filtered['FatCalories'], bar_width, color='r', alpha=0.7, label='Fat Calories')
ax2.set_ylabel('FatCalories', color='r')

ax1.set_xticks([i for i in index])
ax1.set_xticklabels(df_filtered['Menu'], rotation=45, ha='right')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

## กราฟที่ 3 แสดงเปรียบเทียบหมวดหมู่ Chicken ที่มี Cholesterol กับ Fat มากที่สุด
#### ► เหตุผลที่เลือก lineplot เพราะ สามารถเห็นข้อมูลเปรียบเทียบระหว่าง Cholesterol กับ Fat 

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_filtered = df[df['Category'] == 'Chicken'].nlargest(15, 'Fat')

##เลือกรูปแบบของกราฟที่เราจะสร้าง
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

##สร้างกราฟเส้น โดยใช้ Seaborn โดยแกน x คือ "Menu" และแกน y คือ "Fat" และ "Cholesterol" จาก DataFrame 
ax = sns.lineplot(x='Menu', y='Fat', data=df_filtered, label='Fat', marker='o')
ax = sns.lineplot(x='Menu', y='Cholesterol', data=df_filtered, label='Cholesterol', marker='o')

plt.title('Top 10 Chicken Menu Highest Calories and Fat Calories')
plt.xlabel('Menu')
plt.ylabel('Calories / Fat Calories')
plt.xticks(rotation=45, ha='right')
plt.legend()

plt.tight_layout()
plt.show()
```

## กราฟที่ 4 แสดงเมนูที่มี DietaryFiber, Total Carb, Sugars, Protein มากที่สุด 20 อันดับ
#### ► เหตุผลที่เลือก lineplot เพราะ สามารถเห็นความแตกต่างข้อมูลที่ต้องการเปรียบเทียบได้อย่างชัดเจน

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_filtered = df.nlargest(20, ['DietaryFiber','TotalCarb', 'Sugars', 'Protein'])

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(x='Menu', y='DietaryFiber', data=df_filtered, label='DietaryFiber', ci=None)
sns.lineplot(x='Menu', y='TotalCarb', data=df_filtered, label='TotalCarbs', ci=None)
sns.lineplot(x='Menu', y='Sugars', data=df_filtered, label='Sugars', ci=None)
sns.lineplot(x='Menu', y='Protein', data=df_filtered, label='Protein', ci=None)

plt.title('Top 20 Menu with Highest (Fiber, Carbs, Sugars, Protein)')
plt.xlabel('Menu')
plt.ylabel('Values (g)')

plt.legend()
plt.xticks(rotation=90)  

plt.show()
```

## กราฟที่ 5 แสดงเมนูในหมวด Breakfast ที่มี Fat และ Saturated Fat มากที่สุด 12 รายการ
### ► เหตุผลที่เลือก Bar Plot (Multi) เพราะ สามารถแสดงข้อมูล Fat และ Saturated Fat ได้อย่างชัดเจน

```python
plt.style.use('bmh')
import matplotlib.pyplot as plt

df_filtered = df[(df['Category'] == 'Breakfast') & (df['Fat'] > 0)].nlargest(12, 'Fat')

fig, ax1 = plt.subplots(figsize=(12, 12))

bar_width = 0.35
index = range(len(df_filtered))

bar1 = ax1.bar(df_filtered['Menu'], df_filtered['Fat'], bar_width, color='b', alpha=0.7, label='Fat')
ax1.set_ylabel('Fat', color='b')
ax1.tick_params(axis='y', labelcolor='b')
ax1.set_title('Top 12 High-Fat Burgers with Saturated Fat (g)')

ax2 = ax1.twinx()

bar2 = ax2.bar([i + bar_width for i in index], df_filtered['SaturatedFat'], bar_width, color='r', alpha=0.7, label='SaturatedFat')
ax2.set_ylabel('SaturatedFat', color='r')

ax1.set_xticks([i for i in index])
ax1.set_xticklabels(df_filtered['Menu'], rotation=45, ha='right')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
